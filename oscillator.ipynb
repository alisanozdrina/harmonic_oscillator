### Consider a harmonic oscillator of Hamiltonian
$$ H=\frac{1}{2}(p^{2}+\omega^{2}x^2)$$ $$x(0)=0$$ $$p(0)=\omega$$
#### (a) Find the analytic solution.

$$\ddot{x}+\omega^{2}x=0$$
$$x(t)=\sin(\omega t)$$

$$\dot{p}=-\frac{\partial H}{\partial x}$$
$$\dot{p}=-\omega^2 x(t)$$
$$p(t)=\int -\omega^2 x(t) dt = \omega \cos(\omega t)$$

```python
#$\dot{x}=\frac{\partial H}{\partial p}$;  
#$$p(t)=-p(0)\sin(\omega t)+x(0)\omega \cos(\omega t)=-\omega \sin(\omega t)$$
```

```python
import matplotlib.pyplot as plt 
import numpy as np 
t = np.arange(0, 2*(np.pi), 0.1) 
#x = np.sin(t)  
plt.plot(t, np.sin(t), label='$\omega$=1')
plt.plot(t, np.sin(t*2), label='$\omega$=2')
plt.plot(t, np.sin(t*3), label='$\omega$=3')
plt.legend()
plt.legend(fontsize=15)
plt.xlabel('t - axis') 
plt.ylabel('x - axis') 
plt.title('x(t)=$\sin(\omega t)$') 
plt.show() 
```
